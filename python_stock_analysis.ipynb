```python
#股票(抓取、寫入Excel)
#https://www.twse.com.tw/zh/page/trading/exchange/MI_INDEX.html
import pandas as pd
import numpy as np
import csv

import time,os
from selenium.webdriver.chrome.options import Options
from urllib.request import urlopen
from selenium import webdriver
from bs4 import BeautifulSoup

#日期 funtion
def date():
    date = time.localtime(time.time())
    y = int(date.tm_year)-1911
    m = int(date.tm_mon)
    d = int(date.tm_mday)
    return [y,m,d]

# 隱藏瀏覽器
chrome_opts = Options()
#chrome_opts.add_argument("--headless")
driverPath = 'c:/chromedriver.exe'
driver = webdriver.Chrome(options=chrome_opts,executable_path=driverPath)

#每日收盤價 - 台灣證券交易中心資料
#url = #每日收盤價 - 台灣證"https://www.twse.com.tw/zh/page/trading/exchange/MI_INDEX.html"券交易中心資料

url = "https://www.twse.com.tw/zh/page/trading/exchange/MI_INDEX.html"
ori_url = "https://www.twse.com.tw/"
# 開啟瀏覽器`,並隱性等待一秒
driver.get(url)
driver.implicitly_wait(1)

#選擇日期
sp = BeautifulSoup(driver.page_source,"html.parser")
title = sp.select('.logo')
for i in title:
    print(i.text)
    
#選擇年分
ul = driver.find_element_by_xpath("//select[@name='yy']")
ul.find_element_by_xpath('//*[@id="d1"]/select[1]/option[2]').click()
driver.implicitly_wait(1)

#選擇月份
ul = driver.find_element_by_xpath("//select[@name='mm']")
ul.find_element_by_xpath('//*[@id="d1"]/select[2]/option[3]').click()
driver.implicitly_wait(1)

#選擇日期
ul = driver.find_element_by_xpath("//select[@name='dd']")
ul.find_element_by_xpath('//*[@id="d1"]/select[3]/option[9]').click()
driver.implicitly_wait(1)

#選擇股票類型
ul = driver.find_element_by_xpath("//select[@name='type']")
ul.find_element_by_xpath('//*[@id="main-form"]/div/div/form/select/option[21]').click()
driver.implicitly_wait(1)

#選擇全部
ul = driver.find_element_by_xpath('//*[@id="main-form"]/div/div/form/a')
ul.click()
driver.implicitly_wait(1)

# 按下查詢
ul = driver.find_element_by_xpath("//select[@name='report-table1_length']")
ul.find_element_by_xpath('//*[@id="report-table1_length"]/label/select/option[5]').click()
driver.implicitly_wait(1)

# 抓取表格資料匯入CSV
database = [] #全部資料存取區
temp = [] #暫存區

sp = BeautifulSoup(driver.page_source,"html.parser")
tr_ = sp.find_all("tr")

for th in tr_[1]: temp.append(th.text) #存取資料種類
npa = np.array(temp)
database.append(npa)
print(database)
temp.clear()
print(temp)

for tr in tr_:
    td_ = tr.find_all("td")
    if tr_.index(tr) > 1:
        for td in td_:
            temp.append(td.text)
        database.append(temp)
        print(database)
        temp.clear()
        print(temp)

print(database)

#抓110年的今天 至往前 一個月 range(30)
date_today = date()
#for i in range(30):
```
